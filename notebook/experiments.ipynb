```python
from langchain_core.tools import StructuredTool

class WeatherInput(BaseModel):
    city : str

def get_weather(city: str) -> str:
    # Dummy implementation
    return f"The weather in {city} is sunny."
weather_tool = StructuredTool.from_function(
    func=get_weather,
    name="get_weather",
    description="Get the current weather for a given city.",
    input_schema=WeatherInput)
```

```python
class WeatherInput(BaseModel):
    city : str = Field(..., description="The name of the city to get the weather for")
    units : str = Field("metric", description="The units for temperature, either 'metric' or 'imperial'")

class GetWeatherTool(StructuredTool):
    name: ClassVar[str] = "get_weather"
    description: ClassVar[str] = ("Get the current weather for a given city.")
    args_schema: ClassVar[Type[BaseModel]] = WeatherInput

    def _run(self, city: str, units: str = "metric") -> str:
        return get_weather(city,units)
```
